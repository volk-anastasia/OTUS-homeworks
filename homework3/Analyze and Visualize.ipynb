### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
b = np.arange(10,50)
b_inverse = np.flip(b)
b_even = []
for i in b_inverse:
    if i % 2 == 0:
        b_even.append(i)
    
print(b_even)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
c = np.arange(9).reshape(3, 3)
print(c)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
d = np.random.rand(4,3,2)
print(d)
print(f"MIN = {np.min(d)} \nMAX = {np.max(d)}")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
e_1 = np.random.rand(6,4)
e_2 = np.random.rand(4,3)
print(e_1.dot(e_2))
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
f = np.random.rand(7,7)
print("Стандартное отклонение", np.std(f))
print("Среднее отклонение", np.mean(f))
f_norm = f - np.mean(f)/np.std(f)
print(f_norm)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
df = sb.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
df.head(5)
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(f'{df.shape[1]} строчек \n{df.shape[0]} колонок')
```

### Проверьте есть ли пропуски в данных

```python
df_nan = pd.isna(df).sum()
print(df_nan)
print(df_nan.value_counts('0'))
```

### Посмотрите на распределение числовых признаков

```python
print(df[['total_bill', 'tip', 'size']])
```

### Найдите максимальное значение 'total_bill'

```python
max_bill = df['total_bill'].max()
print(max_bill)
```

### Найдите количество курящих людей

```python
print(df.value_counts('smoker').loc['Yes'])
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print(df.day.unique())

df_Sun = df[df['day'] == 'Sun']
print(df_Sun['total_bill'].mean().round(2))

df_Sat = df[df['day'] == 'Sat']
print(df_Sat['total_bill'].mean().round(2))

df_Thur = df[df['day'] == 'Thur']
print(df_Thur['total_bill'].mean().round(2))

df_Fri = df[df['day'] == 'Fri']
print(df_Fri['total_bill'].mean().round(2))
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
tb_median = df['total_bill'].median()
df_median = df[df.total_bill > tb_median]

df_Female = df[df['sex'] == 'Female']
median_F = df_Female['total_bill'].mean().round(2)
print(f"Средняя сумма чаевых у женщин равна {median_F}")

df_Male = df[df['sex'] == 'Male']
median_M = df_Male['total_bill'].mean().round(2)
print(f"Средняя сумма чаевых у мужчин равна {median_M}")
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
df = df.replace(['Yes'], 1)
df = df.replace(['No'], 0)

print(df)
```

# III. Visualization

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
Lunch = list(df[df['time'] == 'Lunch']['tip'])
Dinner = list(df[df['time'] == 'Dinner']['tip'])
sb.histplot(Dinner).set(title='Dinner')
```

```python
sb.histplot(Lunch).set(title='Lunch')
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
df_M = df[df['sex'] == 'Male']
df_F = df[df['sex'] == 'Female']

sb.scatterplot(df_M,
               x=df_M['total_bill'],
               y=df_M['tip'],
               hue='smoker').set(title='Мужчины')
```

```python
sb.scatterplot(df_F,
               x=df_F['total_bill'],
               y=df_F['tip'],
               hue='smoker').set(title='Женщины')
```
